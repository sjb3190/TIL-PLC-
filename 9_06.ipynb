#### Do it! 실습 - mpg 통합 연비 변수 만들기
도시 연비를 의미하는 cty, 고속도로 연비를 의미하는 hwy의 평균을 구해 복합 연비 계산

```python
mpg['total_fuel_efficiency'] = (mpg['cty'] + mpg['hwy'])/2        # 새로운 변수 생성
mpg.head()    # 앞에 5개만 추출
```

```python
mpg['total_fuel_efficiency'].describe()          #통계량 분석
```

히스토램을 만들어 연비 분포를 알아보겠습니다.

```python
mpg['total_fuel_efficiency'].plot.hist()
```

요약 통계량과 히스토그램을 보면
- 복합연비(`total_fuel_efficiency`)의 평균과 중앙값은 약 20임
- 연비가 20~22.5 사이인 자동차 모델이 가장 많음
- 복합연비는 대부분 25 이하이고 25를 넘기는 모델을 많지 않음

```python
mpg = mpg.rename(columns={'total_fuel_efficiency':'total'})    #간단한 변수로 이름바꿈
mpg.head()
```

##  자유자재로 데이터 가공하기
데이터를 분석하려면 데이터를 자유자재로 다룰 수 있어야 합니다. 이 장에서는 데이터를 추출하거나 여러 데이터를 합치는 등의 데이터를 가공하는 방법을 익힙니다.
- 06-1 데이터 전처리 - 원하는 형태로 데이터 가공하기
- 06-2 조건에 맞는 데이터만 추출하기
- 06-3 필요한 변수만 추출하기
- 06-4 순서대로 정렬하기
- 06-5 파생변수 추가하기
- 06-6 집단별로 요약하기
- 06-7 데이터 합치기

|변수명|내용|
|:-:|:-:|
|quert()|행추출|
|df[ ]| 열(변수) 추출|
|sort_values()|정렬|
|groupby()|변수추가|
|assign()|자동차 모델명|
|agg()|통계치 구하기|
|merge()|데이터 합치기 (열)|
|concat() |데이터 합치기 (행)|
